```python
# loading of library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<h3><strong>Providing Future Trends Using The DataSet</strong></h3>

```python
# first visual
#average home price 
plt.figure(figsize=(10,6))
plt.plot(df.index,df["Average Home Price"], color="blue", label="Average Home Price")
plt.title("Average Home Price over time")
plt.xlabel("Date")
plt.ylabel("Price in USD")
plt.grid(True)
plt.legend()
plt.show()
```

```python
#o	Interest rates
#2nd visual 
# a line plot of  interest rate during the same period 
plt.figure(figsize=(10,6))
plt.plot(df.index, df["Interest Rate (%)"], color="green", label="Interest Rate (%)")
plt.title("Interest Rate over time")
plt.xlabel("Date")
plt.ylabel("Interest Rate (%)")
plt.grid(True)
plt.legend()
plt.show()
```
